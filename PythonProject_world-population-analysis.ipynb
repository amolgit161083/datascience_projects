```python
import numpy as np   
import pandas as pd  
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
#Reading the Dataset
df=pd.read_csv('C:/Users/Amol/Desktop/DataEngg/Project/DataSets/WorldPopulation2023.csv')
```

```python
#Checking for the first 10 rolls
df.head(10)
```

```python
#Total Number of Columns and Rows
df.shape
```

```python
#Knowing basic information(dataype,non0-null columns) about the Dataset
df.info()
```

```python
#Dropping duplicate rows if any
df.drop_duplicates()
```

```python
#Check for any Null values
df.isnull().sum()
```

```python
#Drop or replace the Null values
df.dropna()
df.fillna(df['MedianAge'].mean())
```

```python
#To check hidden or non-printable values in the columns.
print(repr(df.columns))
```

```python
#Making Rank the Index of the Dataset
#df.dropna(subset=['Rank'])
df.set_index('Rank',inplace=True)
```

```python
#Sorting countries based on their rank in population
#Top 10 most populated countries
Most_Populated_Countries=df.sort_values(by='Rank',ascending=True).head(10)
```

```python
Most_Populated_Countries.head(10)
```

```python
#using python classify countries on their population basis.

def classify_population(num):
    if num>100000000:
        return('Highly populated')
    
    elif num>1000000:
        return('Moderately populated')
    
    else:
        return('Less populated')
df['population_class']=df['Population2023'].apply(classify_population)
print(df)
df.sort_values(by='population_class',ascending=True).head(10)
```

```python
#Visualization of the top 10 most populated countries
Most_Populated_Countries.plot(kind='bar', x='Country',y='Population2023')
plt.xlabel('Countries')
plt.ylabel('Population2023')
plt.title=('10 MOST POPULATED COUNTRIES')
plt.grid()
plt.show()
```

```python
#10 less populated Countries
Less_Populated=df.sort_values(by='Rank',ascending=True,).tail(10)
Less_Populated
```

```python
#Visualization of10 less populated Countries and their population 
import matplotlib.pyplot as plt

Less_Populated.plot(kind='bar',x='Country',y='Population2023')
plt.title= ('TEN LESS POPULATED COUNTRIES IN 2023')
plt.xlabel('Country'),plt.ylabel('Population2023')
plt.grid()
plt.show()
```

```python
PnL_Correlation =df['Population2023'].corr(df['Land Area(Km²)'])

print("Correlation between Population and Land Area:", PnL_Correlation)
```
